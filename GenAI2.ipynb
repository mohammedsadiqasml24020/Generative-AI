```python
import ollama

prompt = input("Enter your prompt: ")

response = ollama.chat(
    model='llama3.2',
    messages=[
        {
            'role': 'user',
            'content': prompt,
        }
    ]
)

print("\nResponse:\n")
print(response['message']['content'])
```
