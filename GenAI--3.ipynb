```python
import re
import requests
```

```python
SUSPICIOUS_PATTERNS = [
    r"ignore.*instructions",
    r"ignore.*guidelines",
    r"system prompt",
    r"developer message",
    r"reveal.*prompt",
    r"forget.*previous",
    r"bypass",
    r"disable.*safety",
    r"act as",
    r"pretend to be",
]

def check_prompt(prompt):
    prompt = prompt.lower()

    for pattern in SUSPICIOUS_PATTERNS:
        if re.search(pattern, prompt):
            return False, f"Matched pattern: {pattern}"

    return True, "Prompt is safe."
```

```python
OLLAMA_URL = "http://localhost:11434/api/generate"
MODEL = "llama3.2"
```

```python
prompt = input("Enter your prompt: ")

safe, reason = check_prompt(prompt)

if safe:
    print("Guardrail Result: SAFE")
else:
    print("Potential Prompt Injection Detected")
    print("Reason:", reason)
```

```python
if safe:
    payload = {
        "model": MODEL,
        "prompt": prompt,
        "stream": False
    }

    response = requests.post(OLLAMA_URL, json=payload)

    if response.status_code == 200:
        print("\nResponse:\n")
        print(response.json()["response"])
    else:
        print("Error:", response.text)
else:
    print("\nRequest Blocked. This prompt violates AI safety guidelines.")
```
